# Python

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly
import folium
```

```python
df = pd.read_csv("/content/casestudy csv.csv")
```

```python
df.head()
```

```python
df.describe()
```

```python
df.info()
```

```python
df.isnull().sum()
```

```python
columns_to_drop = ['Income', 'Psychosocial Factors']
df = df.drop(columns=columns_to_drop)
```

```python
df
```

```python
sns.countplot(x='GENDER', data=df)
plt.title('Distribution of Gender')
plt.show()
```

```python
sns.countplot(x='Education', data=df)
plt.title('Distribution of Education type')
plt.show()
```

```python
healthcare_visits = df['Frequency of healthcare visits'].value_counts()
plt.figure(figsize=(6, 6))
plt.pie(healthcare_visits, labels=healthcare_visits.index, autopct='%1.1f%%', colors=['orange', 'green'])
plt.title('Distribution of healthcare visits ')
plt.show()
```

```python
plt.figure(figsize=(20, 6))

sns.histplot(df['Sleep duration(Hrs)'], bins=30, kde=True, color="blue")
plt.title('Distribution of Sleep duration(Hrs)')
```

# R

```python
library(dplyr)
library(ggplot2)
```

```python
df <- read.csv("/content/casestudy csv.csv")
```

```python
df
```

```python
ggplot(df, aes(x = Education, fill = GENDER)) +
  geom_bar(position = "stack") +
  labs(title = "Stacked Bar Plot of Education by Gender", x = "Education", y = "Count") +
   theme_minimal() +
  theme(
    plot.title = element_text(size = 20, hjust = 0.5),
    axis.title.x = element_text(size = 20),
    axis.title.y = element_text(size = 20),
    axis.text = element_text(size = 20),
     legend.text = element_text(size = 20),
    legend.title = element_text(size = 20)
  )
```

```python
options(repr.plot.width = 10, repr.plot.height = 6)
ggplot(df, aes(x = Sleep.duration(Hrs), y = Income , color = GENDER)) +
  geom_point() +
  labs(title = "Scatter Plot of Sleep Duration vs Income", x = "Sleep Dura", y = "Reading Score") +
  theme_minimal() +
  theme(
    plot.title = element_text(size = 20, hjust = 0.5),
    axis.title.x = element_text(size = 20),
    axis.title.y = element_text(size = 20),
    axis.text = element_text(size = 20),
     legend.text = element_text(size = 20),
    legend.title = element_text(size = 20)
  )
```

```python
options(repr.plot.width = 10, repr.plot.height = 6)
ggplot(df, aes(x = Sleep.duration.Hrs., y = Income , color = GENDER)) +
  geom_point() +
  labs(title = "Scatter Plot of Sleep Duration vs Income", x = "Sleep Duration", y = "Income") +
  theme_minimal() +
  theme(
    plot.title = element_text(size = 20, hjust = 0.5),
    axis.title.x = element_text(size = 20),
    axis.title.y = element_text(size = 20),
    axis.text = element_text(size = 20),
    legend.text = element_text(size = 20),
    legend.title = element_text(size = 20)
  )
```
